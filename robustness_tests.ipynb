```python
import sys
import os
PATH_TO_DIRECTORY = os.getcwd()
sys.path.append(PATH_TO_DIRECTORY + "/../locomotion_package/")
import locomotion
import json
```

```python
# For each curve, get the paths to their .json files
NUM_CURVES = 50

fr_files = []
res_files = []
# Add in path to robustness test files for each curve
for i in range(NUM_CURVES):
    fr_test_file = PATH_TO_DIRECTORY + "/frame_rate_tests/CRV_" + str(i) + ".json" 
    fr_files.append(fr_test_file)
    
    res_test_file = PATH_TO_DIRECTORY + "/resolution_tests/CRV_" + str(i) + ".json" 
    res_files.append(res_test_file)
```
